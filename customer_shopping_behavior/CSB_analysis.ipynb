```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
df.describe(include='all')    # summary statistics for all columns
```

```python
df.isnull().sum()  # check for missing values in each column
df['Review Rating'] = df['Review Rating'].fillna(df['Review Rating'].median())  # fill missing ratings with median
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
# create a column age_group
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Aged', 'Senior'] 
df['age_group'] = pd.cut(df['age'], bins=[0, 19, 35, 55, 75, 100], labels=labels)
```

```python
df.head()
```

```python
#  create columnn purchase_frequency_days

frequency_mappping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mappping)
df.head()
```

```python
df[['discount_applied','promo_code_used']].head()
```

```python
#  discount_applied matches promo_code_used   so remove promo_code_used

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df.drop(columns=['promo_code_used'], inplace=True)
```
